```python
import numpy as np
import pandas as pd
import os
datapath = 'D:/Workplace/cnn-text-classification-tf/Data/Original_Report.txt'
```

```python
df = pd.read_csv(datapath, sep='|', header=0, encoding='ISO-8859-1', error_bad_lines=False)
```

```python
df
```

```python
#Slicing data and remove null&short text content less than 30 characters
df = df[["MDR_REPORT_KEY", "FOI_TEXT"]]
df = df[(df["FOI_TEXT"].str.len() > 30)]
df = df.drop_duplicates('FOI_TEXT')
```

```python
df_filtered = df[df["FOI_TEXT"].str.contains("|")]
```

```python
df.shape[0]
```

```python
#Combine text by same key
df_combine = df
```

```python
df_combine = pd.DataFrame(df_combine)
df_combine = df_combine.reset_index(level = 0)
```

```python
#Load device problem
probpath = 'D:/Workplace/cnn-text-classification-tf/Data/Report_Device_Problem.txt'
df_prob = pd.read_csv(probpath, sep='|', header=None, names=['MDR_REPORT_KEY', 'PROBLEM_CODE'], encoding='ISO-8859-1', error_bad_lines=False)
df_prob
```

```python
df_final = pd.merge(df_combine, df_prob, how='inner', on='MDR_REPORT_KEY', left_index = True, right_index=False, sort=True, suffixes=('_x', '_y'), copy=True, indicator=False)
```

```python
df_final
```

```python
df_final = df_final.drop_duplicates('FOI_TEXT')
df_final = df_final.dropna()
```

```python
df_final.shape[0]
```

```python
df_final = df_final[df["FOI_TEXT"].str.split(''))==2]
```

```python
df_final.to_csv("input_raw.txt", sep = '|', columns = ['FOI_TEXT', 'PROBLEM_CODE'], header = False, index = False)
```

```python
with open ("input_raw.txt", "r") as rf:
    with open ("input.txt", 'w') as wf:
        for line in rf:
            a = line.split('|')
            if len(a) == 2:
                wf.write(line)
        wf.close
    rf.close
```
